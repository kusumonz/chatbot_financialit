```python
%%writefile financialit.py
import streamlit as st
import os
from google import genai

# ── 1. Konfigurasi Halaman & Estetika Visual (UI/UX) ──────────────────────────
st.set_page_config(page_title="Financialit", page_icon="💸", layout="centered")

# Injeksi CSS Kustom untuk Tema Biru-Hijau (Eye-Catching & Educational)
st.markdown("""
<style>
    /* Mengubah font global ke inter/sans-serif */
    html, body, [data-testid="stSidebarNav"] {
        font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    }
    
    /* Styling tombol utama agar berwarna hijau emerald */
    .stButton>button {
        background-color: #10b981 !important;
        color: white !important;
        border-radius: 8px !important;
        border: none !important;
        transition: all 0.3s ease;
    }
    .stButton>button:hover {
        background-color: #059669 !important;
        transform: translateY(-2px);
        box-shadow: 0 4px 6px rgba(0,0,0,0.1);
    }
    
    /* Custom Styling untuk Sidebar */
    [data-testid="stSidebar"] {
        background-color: #f8fafc;
        border-right: 1px solid #e2e8f0;
    }
</style>
""", unsafe_allow_html=True)

# Banner Utama dengan Gradasi Biru ke Hijau (Teal-Emerald Gradient Card)
st.markdown("""
<div style="background: linear-gradient(135deg, #1e3a8a 0%, #10b981 100%); padding: 25px; border-radius: 15px; color: white; margin-bottom: 25px; box-shadow: 0 10px 15px -3px rgba(0, 0, 0, 0.1);">
    <h1 style="margin: 0; font-size: 26px; font-weight: 700; letter-spacing: -0.5px;">💸 Financialit: Cerdas Finansial & Investasi</h1>
    <p style="margin: 8px 0 0 0; opacity: 0.9; font-size: 14px; line-height: 1.5;">
        Pengen punya financial life yang tertata dan bebas stres? 🤔 Sini bareng asisten pintar kamu, kita bedah strategi atur arus kas sampai cara paling smart buat optimalin portofolio investasimu. No more tebak-tebakan, saatnya kelola duit pakai data! 📊🔥
    </p>
</div>
""", unsafe_allow_html=True)

# Widget Edukatif Dinamis (Info Box dengan Aksen Hijau)
st.markdown("""
<div style="border-left: 5px solid #10b981; background-color: #f0fdf4; padding: 15px; border-radius: 6px; margin-bottom: 20px; box-shadow: 0 1px 2px rgba(0,0,0,0.05);">
    <strong style="color: #1e3a8a; font-size: 15px;">💡 Pilar Keuangan Sehat:</strong> Sebelum melompat ke instrumen investasi berisiko tinggi, pastikan pondasi <em>Cash Flow</em> Anda positif, utang terkendali, dan Dana Darurat telah aman terlindungi. 🛡️
</div>
""", unsafe_allow_html=True)

# ── 2. Manajemen Kredensial & Sidebar (Diperbarui dengan Live Chart) ────────
env_api_key = os.getenv("GEMINI_API_KEY", "")

with st.sidebar:
    # st.markdown("<h2 style='color: #1e3a8a; font-size: 20px; font-weight: 800; letter-spacing: -0.5px;'>🛸 Command Center</h2>", unsafe_allow_html=True)
    
    # Status API dibuat minimalis sebagai indikator saja
    if env_api_key:
        # st.caption("✅ **Engine Secured.** API Key aktif.")
        google_api_key = env_api_key
    else:
        google_api_key = st.text_input("Google AI API Key", type="password")
    
    # st.divider()
    
    st.markdown("<h2 style='color: #1e3a8a; font-size: 16px;'>🎯 Profil Investor</h3>", unsafe_allow_html=True)
    profil_risiko = st.selectbox(
        "Tentukan Profil Risiko Anda:",
        ("Konservatif (Prioritas keamanan modal)", 
         "Moderat (Keseimbangan tren & risiko)", 
         "Agresif (Optimalisasi imbal hasil)")
    )
    
    reset_button = st.button("♻️ Refresh Mesin Analitik", use_container_width=True)
    
    st.divider()

    # === WIDGET LIVE MARKET (IHSG) ===
    st.markdown("<h2 style='color: #1e3a8a; font-size: 16px; margin-bottom: 0px;'>📊 Live Market (IHSG)</h3>", unsafe_allow_html=True)
    
    try:
        import yfinance as yf
        # Menarik data historis IHSG (^JKSE) selama 1 bulan terakhir
        ihsg = yf.Ticker("^JKSE")
        hist = ihsg.history(period="1mo")
        
        if not hist.empty:
            # Mengambil harga penutupan (Close) hari ini dan hari sebelumnya
            last_price = hist['Close'].iloc[-1]
            prev_price = hist['Close'].iloc[-2]
            delta = last_price - prev_price
            
            # 1. Menampilkan angka indeks dan indikator naik/turun otomatis
            st.metric(
                label="Indeks Harga Saham Gabungan", 
                value=f"{last_price:,.2f}", 
                delta=f"{delta:,.2f}"
            )
            
            # 2. Menampilkan grafik area berwarna Emerald Green
            st.area_chart(hist['Close'], color="#10b981", height=150)
            st.caption("Data *real-time* via Yahoo Finance 🌐")
        else:
            st.caption("⏳ Menunggu data market...")
            
    except Exception as e:
        st.warning(f"Gagal memuat chart pasar: {e}")
    # ==================================

    
    
    

# ── 3. Definisi Advanced & Friendly System Instruction ──────────────────────
system_instruction = f"""
Anda adalah 'Financialit', sebuah sistem pakar tingkat lanjut (advanced expert system) dan asisten strategis di bidang Financial Planning, Personal Finance, dan Wealth Management.
Tujuan utama Anda adalah mendidik pengguna agar memiliki literasi keuangan yang komprehensif, mampu merancang arsitektur keuangan pribadi yang tangguh secara empiris, dan membuat keputusan investasi berbasis data.

Saat ini Anda sedang berinteraksi dengan pengguna yang mengklasifikasikan dirinya dengan Profil Risiko: **{profil_risiko}**.

=== PILAR FINANCIAL PLANNING & PERSONAL FINANCE ===
Dalam setiap analisis, Anda WAJIB memetakan situasi pengguna ke dalam hierarki perencanaan keuangan berikut sebelum membahas instrumen investasi:
1. Cash Flow & Debt Management: Evaluasi arus kas masuk/keluar. Terapkan efisiensi alokasi (misal: 50/30/20 rule) dan mitigasi utang (Debt-to-Income Ratio maksimal 30%).
2. Risk Management & Protection: Validasi ketersediaan Dana Darurat (3-12 bulan pengeluaran) dan asuransi (jiwa/kesehatan). Integrasikan optimalisasi jaring pengaman sosial Indonesia seperti BPJS Kesehatan dan BPJS Ketenagakerjaan.
3. Wealth Accumulation (Investasi): Desain alokasi aset yang selaras dengan rentang waktu pencapaian (time horizon) dan profil risiko pengguna.
4. Retirement Planning: Rancang proyeksi masa pensiun menggunakan instrumen seperti DPLK (Dana Pensiun Lembaga Keuangan), reksa dana, atau Surat Berharga Negara (SBN).
5. Tax & Estate Planning: Berikan kesadaran dasar tentang efisiensi perpajakan personal (kepatuhan SPT Tahunan, NPWP) dan distribusi kekayaan.

=== PENDEKATAN ANALITIS & PERSONAL ACCOUNTING ===
- Berpikir layaknya akuntan untuk keuangan pribadi: Arahkan pengguna untuk menyadari pentingnya *Personal Balance Sheet* (Aset vs Liabilitas) guna mengkalkulasi dan melacak *Net Worth* (Kekayaan Bersih).
- Kuantifikasi Analisis: Gunakan rasio kesehatan finansial dalam memberikan saran, seperti Liquidity Ratio (Aset Lancar / Pengeluaran Bulanan), Savings Ratio (target empiris >20%), dan Debt-to-Asset Ratio.
- Behavioral Finance & Nilai Kehidupan: Edukasi pengguna tentang *mindful spending*. Tekankan bahwa keseimbangan finansial juga mencakup harmoni hidup (seperti prinsip Tri Hita Karana—menjaga hubungan baik dengan sesama dan lingkungan). Pengeluaran harus selaras dengan tata nilai (values) yang esensial dan berdampak positif, bukan sekadar pemenuhan gaya hidup impulsif. Identifikasi dan edukasi pengguna mengenai bias kognitif dalam berinvestasi (seperti FOMO, loss aversion, atau overconfidence) jika relevan dengan pertanyaan mereka.
- Grounded Theory: Gunakan kerangka kerja yang mapan seperti Modern Portfolio Theory (MPT), Capital Asset Pricing Model (CAPM), atau pilar dasar Financial Planning (Cashflow Management, Risk Management, Wealth Accumulation, Wealth Distribution).
- Governance, Risk, & Compliance (GRC) Pribadi: Selalu tekankan pentingnya dana darurat, rasio utang (Debt-to-Income Ratio), dan proteksi asuransi sebelum mendiskusikan instrumen investasi risiko tinggi.

=== PARAMETERISASI PROFIL RISIKO ({profil_risiko}) ===
- Jika Konservatif: Prioritaskan capital preservation dan likuiditas tinggi. Rekomendasikan instrumen pasar uang, deposito, SBN Ritel (ORI/Sukuk/SBR), dan mitigasi inflasi.
- Jika Moderat: Terapkan alokasi aset dinamis. Kombinasikan instrumen pendapatan tetap dengan ekuitas atau reksa dana indeks/campuran berfundamental kuat.
- Jika Agresif: Diskusikan optimalisasi imbal hasil melalui porsi ekuitas yang dominan, diversifikasi portofolio, dan pemahaman atas volatilitas. Tetap tekankan money management dan cut-loss/take-profit plan.

=== STRUKTUR & GAYA BAHASA (OUTPUT FORMATTING) ===
1. Tonasi: Profesional, objektif, analitis, logis, sistematis, dan edukatif.
2. Hierarki Informasi: Gunakan heading, bullet points, dan penomoran untuk memecah informasi kompleks menjadi *actionable steps*.
3. Representasi Data: Gunakan format tabel atau rubrik matriks untuk membandingkan instrumen investasi, skenario *budgeting*, atau evaluasi rasio keuangan.
4. Framework Matematis: Jika menjelaskan metrik atau proyeksi nilai waktu uang (Time Value of Money), sertakan simulasi matematis atau rumus komputasinya. Gunakan format LaTeX untuk persamaan, contoh: $FV = PV \\times (1 + r)^n$.
5. Kepribadian: Anda adalah mentor keuangan yang sangat ramah, hangat, penuh semangat (enthusiastic), empati, dan suportif. Anda ingin pengguna merasa nyaman dan tidak takut belajar keuangan.
6. Penggunaan Emoticon: Anda WAJIB menyisipkan emoticons yang relevan di setiap paragraf, sub-judul, atau poin penjelasan (contoh: 💰, 📈, 🚀, 🛡️, 📊, 📝, 😊, 👍) untuk menjaga interaksi tetap menarik secara visual dan menyenangkan.
7. Kejelasan: Gunakan analogi sederhana untuk menjelaskan istilah rumit agar mudah dipahami pembelajar awam.

=== STRICT COMPLIANCE & GUARDRAILS ===
1. ZERO Financial Advice: Anda adalah platform EDUKASI literasi keuangan, bukan Certified Financial Planner (CFP) yang memberikan rekomendasi buy/sell spesifik. Selalu akhiri analisis dengan disclaimer bahwa eksekusi keputusan berada di tangan pengguna.
2. Anti-Halusinasi & Realisme: Jangan memprediksi pergerakan pasar secara absolut. Gunakan data historis, probabilitas, dan analisis makroekonomi untuk menjelaskan skenario. Fokus pada instrumen dan ekosistem keuangan Indonesia (OJK, BEI, LPS).
"""

# ── 4. Inisialisasi State & Gemini Client ─────────────────────────────────────
if ("genai_client" not in st.session_state) or \
   (getattr(st.session_state, "_last_key", None) != google_api_key) or \
   (getattr(st.session_state, "_last_profile", None) != profil_risiko):
    
    try:
        st.session_state.genai_client = genai.Client(api_key=google_api_key)
        st.session_state._last_key = google_api_key
        st.session_state._last_profile = profil_risiko
        
        st.session_state.pop("chat", None)
        st.session_state.pop("messages", None)
    except Exception as e:
        st.error(f"Kegagalan otentikasi API Key: {e}")
        st.stop()

if "chat" not in st.session_state:
    st.session_state.chat = st.session_state.genai_client.chats.create(
        model="gemini-2.5-flash",
        config={"system_instruction": system_instruction, "temperature": 0.4}
    )

if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "assistant", "content": f"Halo Sahabat Finansial! 👋 Selamat datang di Financialit! 😊 Saya sangat senang bisa mendampingi proses belajar Anda hari ini. Berdasarkan profil pilihan Anda, kita akan menyusun strategi keuangan yang cocok untuk tipe **{profil_risiko.split(' ')[0]}**! 🎯 Pintu diskusi sudah terbuka, apa yang ingin kita bedah bersama hari ini? 💰📈"}
    ]

if reset_button:
    st.session_state.pop("chat", None)
    st.session_state.pop("messages", None)
    st.rerun()

# ── 5. Render Jalur Percakapan (Memory) ────────────────────────────────────
for msg in st.session_state.messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

# ── 6. Logika Input & Pemrosesan Respons AI ──────────────────────────────────
prompt = st.chat_input("Tulis pertanyaan Anda di sini... (Contoh: Bagaimana cara bagi gaji 5 juta?)")

if prompt:
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)

    with st.chat_message("assistant"):
        message_placeholder = st.empty()
        try:
            response = st.session_state.chat.send_message(prompt)
            answer = response.text if hasattr(response, "text") else str(response)
            message_placeholder.markdown(answer)
        except Exception as e:
            answer = f"⚠️ **Aduh, maaf sekali terjadi kendala teknis pada sistem:** {e}. Mari coba segarkan halaman atau periksa koneksi Anda ya! ✨"
            message_placeholder.markdown(answer)
            
    st.session_state.messages.append({"role": "assistant", "content": answer})
```

```python
import os
import subprocess
import time
from pyngrok import ngrok
from google.colab import userdata

def launch_financialit(filename="financialit.py", port=8501):
    try:
        # Menarik data dari menu Secrets (Ikon Kunci 🔑)
        ngrok_token = userdata.get('NGROK_TOKEN')
        gemini_api_key = userdata.get('GEMINI_API_KEY')
    except userdata.SecretNotFoundError as e:
        print(f"❌ Kredensial Tidak Ditemukan: {e}")
        print("Silakan simpan NGROK_TOKEN dan GEMINI_API_KEY di panel Secrets terlebih dahulu.")
        return None

    # Pengaturan parameter keamanan & sub-proses
    ngrok.set_auth_token(ngrok_token)
    os.environ["GEMINI_API_KEY"] = gemini_api_key

    # Pembersihan port untuk mencegah tabrakan jaringan
    subprocess.run(["pkill", "-f", "streamlit"], capture_output=True)
    subprocess.run(["fuser", "-k", f"{port}/tcp"], capture_output=True)
    ngrok.kill()
    time.sleep(2)

    # Eksekusi server internal Streamlit
    proc = subprocess.Popen(
        ["streamlit", "run", filename, "--server.headless=true", "--server.port", str(port)],
        stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL
    )
    time.sleep(3)

    # Pembuatan tautan publik aman (Secure Tunnelling)
    public_url = ngrok.connect(port)
    print(f"==========================================================")
    print(f"🔒 Otentikasi Berhasil. Kredensial Tersimpan Aman di Memory.")
    print(f"🚀 Financialit Dashboard Siap Digunakan!")
    print(f"🔗 Klik tautan ini untuk membuka Chatbot Anda: {public_url}")
    print(f"==========================================================")
    
    return proc

# Eksekusi fungsi peluncuran
proc = launch_financialit()
```
